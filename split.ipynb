```python
from PIL import Image
import os
import numpy as np
import matplotlib.pyplot as plt

path = r'D:\pic'

im = Image.open(os.path.join(path, 'cut2.jpg'))

w, h = im.size

cols = 12
rows = 14
wcell = w / cols
hcell = h / rows

stroke = 5

splits = []
for i in range(rows):
    for j in range(cols):
        newimg = im.crop((
            j * wcell + stroke,
            i * hcell + stroke,
            (j + 1) * wcell - stroke,
            (i + 1) * hcell - stroke))
        splits.append(np.array(newimg)[:, :, 0])
        newimg.save(os.path.join(path, 'splits\%(i)d-%(j)d.jpg' % {'i': i, 'j': j}))
splits = np.array(splits).reshape((rows, cols))
```

```python
for i in range(cols - 1):
    plt.imshow(splits[i][i], cmap=plt.cm.gray)
    plt.show()
```
